```python
import numpy as np
import pandas as pd
import plotly.express as px
import matplotlib.pyplot as plt
from scipy.ndimage import rotate
from sklearn.model_selection import train_test_split
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.neural_network import MLPClassifier
from sklearn.linear_model import LinearRegression
from sklearn.metrics import classification_report, confusion_matrix
import torchvision
```

```python
# import the dataset

dataset = torchvision.datasets.FashionMNIST(root="./data", train=True, download=True)
images = dataset.data.numpy().reshape(-1, 784).astype(float)
labels = np.array([dataset.classes[t] for t in dataset.targets.numpy()])

df = pd.DataFrame({"label": labels})
df['image'] = list(images)
```

```python
# check repartition
df["label"].value_counts()
```

```python
#dataset is balanced, lets split the data
X_train,X_test, y_train, y_test = train_test_split(df["image"],df["label"],test_size = 0.2)
df_train = pd.DataFrame({"image":X_train,"label":y_train})
```

```python
df_train
```

```python
#train Kmeans
classifier = KMeans(n_clusters = 10)
predictions = classifier.fit_predict(np.stack(X_train.values),y_train.values)
predictions
```

```python
#add the cluster to the df
df_train["cluster"] = predictions
```

```python
repartition_cluster = df_train.groupby("label")["cluster"].value_counts().reset_index()
```

```python
repartition_cluster
```

```python

pd.options.plotting.backend = "plotly"
repartition_cluster.plot.bar(x = "cluster",y="count",color = "label")
```
